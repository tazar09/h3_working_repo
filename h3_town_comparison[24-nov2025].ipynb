```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://heroes.thelazy.net/index.php/Town_Comparison'
```

```python
df0 = pd.read_html(url)[0]
```

```python
df = df0.copy()
```

```python
df = df.droplevel(level = 0, axis =1)
df = df.drop(0, axis = 0)
df = df.drop(11, axis = 0)
```

```python
df.head()
```

```python
df.columns = ['town', 'mage_guild', 'silo', 'war_machine', 'week_gold',
       'week_res', 'dwell_ai_val', 'dwell_fight_val', 'cost_gold', 'cost_wood_ore',
       'cost_res', 'wood', 'mercury',  'ore', 'sulf', 'crystal','gems']
df = df.drop(['war_machine'], axis = 1)
```

```python
df.head()
```

```python
resource_silo = {'1 , 1': 'w/o', '1': 'res'}
```

```python
df['silo'] = df['silo'].replace(resource_silo)
df['mage_guild'] = df['mage_guild'].str.replace('[^\d\.]+', '', regex=True)
```

```python
df['week_gold'] = df['week_gold'].apply(lambda x: str.split(x)[0])
df['week_res'] = df['week_res'].apply(lambda x: str.split(x)[0])
df['dwell_ai_val'] = df['dwell_ai_val'].apply(lambda x: str.split(x)[0])
df['cost_gold'] = df['cost_gold'].apply(lambda x: str.split(x)[0])
df['cost_wood_ore'] = df['cost_wood_ore'].apply(lambda x: str.split(x)[0])
df['cost_res'] = df['cost_res'].apply(lambda x: str.split(x)[0])
df['mercury'] = df['mercury'].apply(lambda x: str.split(x)[0])
df['ore'] = df['ore'].apply(lambda x: str.split(x)[0])
df['sulf'] = df['sulf'].apply(lambda x: str.split(x)[0])
df['crystal'] = df['crystal'].apply(lambda x: str.split(x)[0])
```

```python
df.columns
```

```python
df.head(20)
```

```python
df = df.drop(index = 12)
for i in ['mage_guild', 'week_gold', 'week_res', 'dwell_ai_val', 'cost_gold', 'cost_wood_ore', 'cost_res', 'wood','mercury', 'ore', 'sulf', 'crystal', 'gems']:
  df[i] = df[i].astype(int)
```

```python
df['cost_ai'] = round(df['week_gold']/df['dwell_ai_val'],2)
```

```python
# df.sort_values('cost_res_std', ascending = False)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['cost_res_std'] = scaler.fit_transform(df['cost_res'].values.reshape(-1, 1)).round(2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['cost_wood_ore_std'] = scaler.fit_transform(df['cost_wood_ore'].values.reshape(-1, 1)).round(2)
```

```python
df.sort_values('mercury', ascending=False)
```

Trying to convert different resources to the same standard:

```python
lst = []
res = ['mercury', 'sulf', 'crystal', 'gems']


df = df.drop(index = [9,10])
for i in ('mercury', 'sulf', 'crystal', 'gems'):
  lst.append(df[i].sum())
  x = df[i].sum()
  print(i,":", x)

print(sum(lst)/4)
```

```python
exchange_rate = {}
for i,j in zip(lst,res):
  print(j, ':', round(i/(sum(lst)/4),2))
  exchange_rate[j] = round(i/(sum(lst)/4),2)
print(exchange_rate)
```

```python
for i in res:
  df[i] = df[i].apply(lambda x: round(x*exchange_rate[i],1))
df.sort_values('mercury', ascending=False)
```
